# Data Featuring/Categorical variable 

```python
df.select_dtypes("object") # These are all the categorical columns we need to change it to 
                           #numerical by labeling or OnHotEncoding
```

```python
# As we can analyse from the above table that Item_Identifier and Item_Type have the connection with the names
# we can consider Dairy comes under 'FD' which will be 'Food' and 'DR'-'Soft Drinks' and 'NC'--HouseHold
df.Item_Identifier.value_counts()
```

```python
df['Item_Identifier'] = df['Item_Identifier'].apply(lambda x: x[0:2])
```

```python
df.Item_Identifier.value_counts()
```

```python
df['Item_Identifier']=df['Item_Identifier'].map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})
```

```python
df.Outlet_Location_Type.value_counts()
```

```python
df['Outlet_Location_Type'] = df['Outlet_Location_Type'].apply(lambda x: x[-1])
```

```python
df.Outlet_Location_Type.value_counts()
```

```python
df.Outlet_Type.value_counts()
```

```python
columns=df.select_dtypes("object").columns
for features in columns:
    print(features,":",len(df[features].unique()),"labels")
```

```python
#from the above we can say that we have to reduce the labels for Item_Type , Item_Fat_Content, and Outlet_Identifier
df.Item_Fat_Content.unique()
```

```python
df.Item_Fat_Content.replace(['LF'], "low fat",inplace=True)
df.Item_Fat_Content.replace(['reg'], "Regular",inplace=True) 
```

```python
df.Item_Type.unique()
```

```python
df.Item_Type.replace(['Canned','Snack Foods','Breads','Soft Drinks','Hard Drinks','Baking Goods'], "Snack",inplace=True)
df.Item_Type.replace(['Starchy Foods','Fruits and Vegetables'],'Fruit&Veg',inplace=True)
df.Item_Type.replace(['Health and Hygiene','Household'],'House_Hold',inplace=True)
df.Item_Type.replace(['Meat','Frozen Foods','Seafood'],'Frozen_items',inplace=True)
df.Item_Type.replace(['Dairy','others'],'Dairy',inplace=True)
```

```python
df.Item_Type.unique()
```

```python
columns=df.select_dtypes("object").columns
for features in columns:
    print(features,":",len(df[features].unique()),"labels")
```

```python
df['Outlet_Identifier'] = df['Outlet_Identifier'].apply(lambda x: x[-2:])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le_columns=['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Type']
df[le_columns] = df[le_columns].apply(lambda col: le.fit_transform(col)) 
```

```python
# Now We have changed every categorical column into numerical
df
```

# Model Building

```python
#Divide the df into test and train:
train = df.loc[df['data']=="train_dataset"]
test = df.loc[df['data']=="test_dataset"]
train.drop(columns=['data'],inplace=True)
test.drop(columns=['data'],inplace=True)
```

```python
train.shape,test.shape
```

```python
train.isna().sum()
```

```python
x_train=train.drop(columns=["Item_Outlet_Sales"])
y_train=train.Item_Outlet_Sales
```

```python
x_test=test.drop(columns=["Item_Outlet_Sales"])
```

```python
x_test.shape,x_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,accuracy_score,r2_score
r2_score(y_train, model.predict(x_train))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
y_dtr_pred=dtr.predict(x_test)
```

```python
r2_score(y_train, dtr.predict(x_train))
```
